As it is taking very long time to traslate , I am taking only 5000 unique record

```python
df_unique=df.drop_duplicates(keep='first', inplace=False)
df_unique=df_unique[:5000]
df_unique
```

```python
# 'es': 'spanish'

spanish=[]
for i in df_unique["English"]:
  translator=Translator()
  j=translator.translate(i, dest="es")
  spanish.append(j.text)
 
```

```python
#'fr': 'french'

french=[]
for i in df_unique["English"]:
  translator=Translator()
  k=translator.translate(i, dest="fr")
  french.append(k.text)
 
```

```python
#'de': 'german'

german=[]
for i in df_unique["English"]:
  translator=Translator()
  l=translator.translate(i, dest="de")
  german.append(l.text)
  
```

```python
print(spanish,french,german,len(spanish),len(french),len(german), sep="\n\n")
```

```python
result=pd.DataFrame(list(zip(df_unique["English"],spanish,french,german)),columns=["English","Spanish","French","German"])
result
```
